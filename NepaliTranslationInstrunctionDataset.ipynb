```python
from datasets import load_dataset

# Load the dataset from Hugging Face
dataset = load_dataset("ashokpoudel/English-Nepali-Translation-Instruction-Dataset")
```

```python
# Assuming the dataset is in the format as mentioned in your example:
# <s>[INST] Please translate "..." into Nepali [/INST] Sure! here is the translated text into Nepali: ... </s>
# We will extract the English and Nepali sentences.

def extract_translations(example):
    # Extract the English sentence from within the quotes
    en_start = example['text'].find('translate "') + len('translate "')
    en_end = example['text'].find('"', en_start)
    english_sentence = example['text'][en_start:en_end].strip()

    # Extract the Nepali sentence after the Nepali translation indicator
    np_start = example['text'].find('Nepali:') + len('Nepali: ')
    np_end = example['text'].find('</s>', np_start)
    nepali_sentence = example['text'][np_start:np_end].strip()

    # Return the extracted sentences as a dictionary
    return {'en': english_sentence, 'np': nepali_sentence}
```

```python
# Apply the function to each dataset split
translated_dataset = dataset.map(extract_translations, batched=False)
```

```python
from datasets import load_dataset, Dataset, DatasetDict

# Load data
with open('/content/corpus_NOV2019/corpus_NOV2019/collected/train.en', 'r', encoding='utf-8') as en_file, \
     open('/content/corpus_NOV2019/corpus_NOV2019/collected/train.ne', 'r', encoding='utf-8') as ne_file:
    english = en_file.readlines()
    nepali = ne_file.readlines()

# Convert to the desired format for English to Nepali
en_to_ne_data = [f'<s>[INST] Please translate "{en.strip()}" into Nepali [/INST] Sure! here is the translated text into Nepali: {ne.strip()} </s>' for en, ne in zip(english, nepali)]

# Convert to the desired format for Nepali to English
ne_to_en_data = [f'<s>[INST] Please translate "{ne.strip()}" into English [/INST] Sure! here is the translated text into English: {en.strip()} </s>' for en, ne in zip(english, nepali)]

# Concatenate the two lists
combined_data = en_to_ne_data + ne_to_en_data

# Create a Dataset with the combined data
dataset = Dataset.from_dict({"text": combined_data})

# Push dataset to Hugging Face
dataset.push_to_hub("English-Nepali-Translation-Instruction-Dataset")
```
